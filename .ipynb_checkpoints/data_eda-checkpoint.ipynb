# Variation of the price range by the vehicle type

```python
# Boxplot to see the distribution after outliers has been removed
def remove_outlier(df_in, col_name):
    q1 = df_in[col_name].quantile(0.25)
    q3 = df_in[col_name].quantile(0.75)
    iqr = q3-q1 #Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    df_out = df_in.loc[(df_in[col_name] > fence_low) & (df_in[col_name] < fence_high)]
    return df_out
df1 = remove_outlier(df,'Price')

sns.set_style("whitegrid")
fig, ax = plt.subplots(figsize=(17,10))
sns.boxplot(x="Price", y="Category", data=df1)
ax.text(10000,-1,"Boxplot After removing outliers",fontsize=18,color="r",ha="center", va="center")
ax.xaxis.set_label_text("Price Range",fontdict= {'size':14})
ax.yaxis.set_label_text("Type Of Vehicle",fontdict= {'size':14})
plt.show()
```

```python
# saving the plot
fig.savefig('Images/price-vehicleType-boxplot.png')
```

# HeatMap of vehicles by Zip_codes over the Greek Map

```python
Zip_codes=df.Zip_code.unique()
df_geocoders = pd.DataFrame(data=Zip_codes,columns=['Zip_code'])


df_geocoders['Latitude'] = None
df_geocoders['Longitude'] = None

locator = Nominatim(user_agent="myGeocoder")
   
for i in range(0, len(df_geocoders), 1):
    location = locator.geocode({"postalcode": str(df_geocoders.iat[i,0]), 'country': 'Ελλάδα'})
    try:
        lat = location.latitude
        lon = location.longitude
        df_geocoders.iat[i,df_geocoders.columns.get_loc("Latitude")] = lat
        df_geocoders.iat[i,df_geocoders.columns.get_loc("Longitude")] = lon
    except:
        lat = None
        lon = None
    
df = df.merge(df_geocoders, left_on='Zip_code', right_on='Zip_code')
```

```python
m = folium.Map(location=[39.074208,21.824312] , tiles="Stamen Terrain" ,zoom_start = 6)
df['Latitude'] = df['Latitude'].astype(float)
df['Longitude'] = df['Longitude'].astype(float)

heat_df = df[['Latitude', 'Longitude']]
heat_df = heat_df.dropna(axis=0, subset=['Latitude','Longitude'])

# List comprehension to make out list of lists
heat_data = [[row['Latitude'],row['Longitude']] for index, row in heat_df.iterrows()]

# Plot it on the map
HeatMap(heat_data).add_to(m)

# Display the map
m
```
